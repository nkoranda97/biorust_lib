```python
sequences = read_csv(
    "/Users/nick/Library/CloudStorage/Box-Box/Zhang/projects/tf_cloning/Edited-FinalTF sheet.csv",
    id_col="DNASU_Clone_ID",
    seq_col="Sequence",
    on_error="skip",
)
```

```python
translations = sequences.seqs().translate()
```

```python
sum(translations.lengths()) / len(translations)
```

```python
highest = 0.0
highest_result: tuple[int | None, int | None] = None, None
scoring = Scoring(gap_open=-10, gap_extend=-0.5)

for i1, pro1 in enumerate(translations):
    for i2, pro2 in enumerate(translations):
        if pro1 == pro2:
            pass

        results = align_global(pro1, pro2, scoring=scoring, traceback=False)
        if results.score > highest:
            highest = results.score
            print(highest)
            highest_result = i1, i2


print(highest_result)
```
